```python
HR.salary = HR.salary.map({'low':0,'medium':1,'high':2})
```

```python
X = HR.drop(['left','role'],1)
y = HR['left']
```

```python
from sklearn.preprocessing import StandardScaler
scaled_features = StandardScaler().fit_transform(X.values)
```

```python
new_df3 = pd.DataFrame(scaled_features, columns = X.columns)
```

```python
import numpy as np
from sklearn import svm
```

```python
#splitting the train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test= train_test_split(new_df3,y,test_size=0.2)
#Trainng and fitting data into the model classifier
clf = svm.SVC()
clf.fit(X_train,y_train)
```

```python
#Accuracy of sVM classifier
accuracy = clf.score(X_test,y_test)
print(accuracy *100,'%')

#Displaying some of the predicted values
pd.DataFrame(clf.predict(X_test),y_test,columns=['ytest']).head()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification
from sklearn.ensemble import ExtraTreesClassifier


# Build a forest and compute the feature importances
forest = ExtraTreesClassifier(n_estimators=250,
                              random_state=0)

forest.fit(X_train, y_train)
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

col_name=list(X)
for f in range(X.shape[1]):
    print("%d.   %s (%f)"  % (f + 1, col_name[f], importances[indices[f]]))

# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(X.shape[1]), importances[indices],
       color="r",width= 0.8, yerr=std[indices], align="center")
plt.xticks(range(X.shape[1]), col_name, rotation='vertical')
plt.xlim([-1, X.shape[1]])
plt.show()
```
